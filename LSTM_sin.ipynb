```python
from chainer import Chain
from chainer import links as L
from chainer import functions as F
from chainer import optimizers
from chainer import Variable
```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def var(x):
    return Variable(np.array(x).astype(np.float32))

class RNN(Chain):
    def __init__(self):
        super(RNN, self).__init__()
        with self.init_scope():
            self.fc1 = L.Linear(1, 50)  # word embedding
            self.lstm1 = L.LSTM(50, 20)  # the first LSTM layer
            self.fc2 = L.Linear(20, 1)  # the feed-forward output layer

    def reset_state(self):
        self.lstm1.reset_state()

    def __call__(self, x):
        # Given the current word ID, predict the next word.
        x = self.fc1(x)
        x = F.relu(x)
        x = self.lstm1(x)
        x = F.relu(x)
        y = self.fc2(x)
        return y

def create_data(n_data,dt,length):
    data=[]
    for k in range(n_data):
        sta = np.random.rand() * 2*np.pi
        en  = sta + length
        d = np.arange(sta,en,dt)
        data.append(d)
    y = np.sin(np.vstack(data))
    y += np.random.randn(y.shape[0],y.shape[1])#noise added
    return y
```

```python
np.random.seed(1)
```

```python
#data setting
dt = np.pi/5
length = 20
seqlen=int(length/dt)
```

```python
y=create_data(3,dt,length)
for k in range(3):
    plt.plot(y[k])
plt.title('data_sample')
plt.savefig('data_example.png')
```

```python
opt = optimizers.SGD()
model=RNN()
opt.setup(model)
```

```python
print('start train')
ite = 300
batch_size = 10
for k in range(ite):
    model.reset_state()
    loss = 0
    count=0

    y=create_data(batch_size,dt,length)
    y=var(y)

    for t in range(1,seqlen):
        p = model(y[:,t-1:t])
        loss += (p[0][0]-y[0,t])**2
        count+=1
        if count % 10 == 0 or count == seqlen:
            model.cleargrads()
            loss.backward()
            loss.unchain_backward()
            opt.update()
print('end train')
```

```python
def prediction_from_scalar(model,start):
    #prediction from scalar using RNN model
    pred=[]
    p = model(start)
    pred.append(p.data)
    for t in range(seqlen-1):
        p = model(p)
        pred.append(p.data)
    return pred
    
```

```python
sta=0
x = var([sta]).reshape(-1,1)
pred = prediction_from_scalar(model,x)
plt.plot(np.vstack(pred))
plt.title('predicted from scalar using RNN')
plt.savefig('predicted_from_scalar_using_RNN.png')
```
